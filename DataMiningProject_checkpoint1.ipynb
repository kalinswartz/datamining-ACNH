# D) Exploratory Data Analysis

```python
import pandas as pd
import kagglehub
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
import itertools

# Get dataset and load into dataframe
path = kagglehub.dataset_download("jessicali9530/animal-crossing-new-horizons-nookplaza-dataset")
villagers = pd.read_csv(path + "/villagers.csv")

display(villagers)

# Convert DD-MMM to datetime (add dummy year 2000) for future ease of use
villagers['Birthday'] = pd.to_datetime(villagers['Birthday'] + '-2000', format='%d-%b-%Y', errors='coerce')

#Check if formatted correctly
villagers.shape
villagers.info()
villagers.head()

# Get frequency of categorical items
categorical_cols_graph = ['Species', 'Gender', 'Personality', 'Hobby', 'Style 1', 'Style 2', 'Color 1', 'Color 2']
categorical_cols_list = ['Favorite Song', 'Wallpaper', 'Flooring']

#Plot distributions of categorical items
for col in categorical_cols_graph:
    plt.figure(figsize=(8,5))
    sns.countplot(y=col, data=villagers, order=villagers[col].value_counts().index)
    plt.title(f"Distribution of {col}")
    plt.show()

#Print out lists of categorial cols with many unique values
for col in categorical_cols_list:
    print(f"--- {col} ---")
    print(villagers[col].value_counts())
    print("\n")

#Plot birthday month distribution
villagers['birth_month'] = villagers['Birthday'].dt.month
sns.countplot(x='birth_month', data=villagers)
plt.title("Villager Birthdays by Month")
plt.show()

# --- Furniture List Analysis ---
#Split the Furniture List into lists of IDs
villagers['furniture_ids'] = villagers['Furniture List'].fillna("").apply(lambda x: x.split(';'))

#List of CSV files containing furniture items
csv_files = [path + "/housewares.csv",
             path + "/miscellaneous.csv",
             path + "/wall-mounted.csv"
            ]  # add more files as needed

#Build master ID -> Name mapping
mapping_dfs = []
for f in csv_files:
    df = pd.read_csv(f, usecols=['Internal ID', 'Name'])  # keep only relevant columns
    df= df.dropna()
    # Convert to numeric first, then to string
    df['Internal ID'] = pd.to_numeric(df['Internal ID'], errors='coerce')
    df['Internal ID'] = df['Internal ID'].astype(int).astype(str)

    df['Name'] = df['Name'].astype(str).str.strip()

    mapping_dfs.append(df)

#Combine all into one dataframe
master_mapping_df = pd.concat(mapping_dfs, ignore_index=True)
master_mapping_df = master_mapping_df.drop_duplicates(subset='Internal ID')

# Create dictionary for fast mapping
id_to_name = dict(zip(master_mapping_df['Internal ID'].astype(str), master_mapping_df['Name']))

#Map furniture IDs to names using master mapping
def map_furniture_id(furniture_id):
    furniture_id = furniture_id.strip()
    return id_to_name.get(furniture_id, furniture_id)  # keep original ID if not found

villagers['furniture_names'] = villagers['furniture_ids'].apply(lambda lst: [map_furniture_id(i) for i in lst])

#Flatten all items and count frequency
all_items = [item for sublist in villagers['furniture_names'] for item in sublist if item]
item_counts = Counter(all_items)

# Top 20 furniture items
top_items = item_counts.most_common(20)
top_items_df = pd.DataFrame(top_items, columns=['Item', 'Frequency'])
display(top_items_df)

# Basket Size Distribution
villagers['basket_size'] = villagers['furniture_names'].apply(len)

plt.figure(figsize=(8,5))
sns.histplot(villagers['basket_size'], bins=20)
plt.title("Distribution of Furniture Basket Sizes")
plt.xlabel("Number of Items")
plt.ylabel("Number of Villagers")
plt.show()

print("Average basket size:", villagers['basket_size'].mean())

# Co-occurence Sparsity of furniture
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
item_matrix = pd.DataFrame(
    mlb.fit_transform(villagers['furniture_names']),
    columns=mlb.classes_
)

sparsity = 1 - (item_matrix.values.sum() / (item_matrix.shape[0] * item_matrix.shape[1]))
print(f"Sparsity of item matrix: {sparsity:.4f}")

# Pairwise Co-occurence matrix
pair_counts = Counter()

for basket in villagers['furniture_names']:
    for pair in itertools.combinations(sorted(set(basket)), 2):
        pair_counts[pair] += 1

top_pairs = pair_counts.most_common(10)
display(pd.DataFrame(top_pairs, columns=["Item Pair", "Count"]))

# Personality vs Species
display(pd.crosstab(villagers['Personality'], villagers['Species']))

# Personality vs Hobby
display(pd.crosstab(villagers['Personality'], villagers['Hobby']))

# Favorite Song vs Personality
song_vs_personality = pd.crosstab(villagers['Personality'], villagers['Favorite Song'])
display(song_vs_personality)

# Run chi square test on favorite song vs personality
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(song_vs_personality)

print("Favorite Song vs Personality Chi Squared Test Stats:")
print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("P-value:", p)
```

An exploratory data analysis (EDA) was done on the dataset. Multiple metrics were outputted, such as distributions of the categorical variables: species, gender, personality, hobby, style 1/2, color 1/2, favorite song, wallpaper, and flooring. Then the villager birthdays were sorted and counted by month to see any temporal patterns. Then an analysis of the furniture ID lists of each villager was done by co-referencing other datasets including the name and ID of different furniture items. A mapping was creating between the Names and ID of furniture items to convert the ID furniture lists into named furniture lists - then the top 20 frequent furniture items was found. I also took a look at the basket size distributions of the furniture lists. I calculated the co-occurence sparsity, and also looked at a pairwise co-occurence matrix of furniture items. Additional cross-analysis metrics were explored such as personality vs species, personality vs hobby, and favorite song vs personality to get ideas for future analysis. Finally, I ran a chi-squared test on favorite song vs personality to check for associations. Because of the many columns of data in this villager dataset, many more permutations and analyses are possible outside of this initial EDA.
